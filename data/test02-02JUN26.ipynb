## Questão 8 — Sistema Paralelo 1oo3 com Falhas de Causa Comum (CCF)

Considere um sistema em paralelo na configuração 1x3 (três componentes em paralelo sendo necessário o funcionamento de apenas um para o cumprimento de sua missão). Sabe-se que, durante um teste, para realização de sua missão de 2000 horas, foi verificada a ocorrência de 45 eventos de falha de um único componente, 1 evento de falha simultânea de 2 componentes e 1 evento de falha simultânea de 3 componentes. A partir destas informações determine:

1. a probabilidade total de falha de um componente.
2. a probabilidade de falha independente de um componente.
3. a probabilidade de falha de mais de um componente simultaneamente.
4. a probabilidade de falha do sistema.

---

### Modelo de Solução

Este problema emprega o **Modelo dos Fatores Alpha (α)** para falhas de causa comum (CCF).

Adota-se $N = 2000$ como o número total de demandas (missões de referência), onde:

| Símbolo | Descrição | Valor |
|---------|-----------|-------|
| $m$ | Número de componentes | 3 |
| $N$ | Total de demandas (horas de missão) | 2000 |
| $n_1$ | Eventos com exatamente 1 componente em falha | 45 |
| $n_2$ | Eventos com exatamente 2 componentes em falha simultânea | 1 |
| $n_3$ | Eventos com exatamente 3 componentes em falha simultânea | 1 |

O total de falhas de componentes (contando multiplicidades) é:
$$\sum_k k \cdot n_k = 1\cdot45 + 2\cdot1 + 3\cdot1 = 50$$

#### 1. Probabilidade Total de Falha de um Componente ($Q_t$)

Divide-se o total de falhas de componentes pelo número total de oportunidades ($m \times N$):

$$Q_t = \frac{\sum_k k \cdot n_k}{m \cdot N} = \frac{50}{3 \times 2000} = \frac{50}{6000} = \frac{1}{120} \approx 8{,}33 \times 10^{-3}$$

#### 2. Probabilidade de Falha Independente de um Componente ($Q_i$)

Falhas independentes são aquelas em que **apenas 1 componente** falha (eventos $n_1$). Por simetria, cada componente é responsável por $n_1/m$ desses eventos:

$$Q_i = \frac{n_1}{m \cdot N} = \frac{45}{3 \times 2000} = \frac{45}{6000} = \frac{3}{400} = 7{,}5 \times 10^{-3}$$

Verificação: $Q_t = Q_i + Q_{i,\text{CCF}}$, onde $Q_{i,\text{CCF}} = \frac{2n_2/3 + n_3}{N} = \frac{5/3}{2000} = \frac{5}{6000} \approx 8{,}33 \times 10^{-4}$ ✓

#### 3. Probabilidade de Falha Simultânea de Mais de um Componente ($Q_{\text{CCF}}$)

São os eventos em que **2 ou mais** componentes falham ao mesmo tempo:

$$Q_{\text{CCF}} = \frac{n_2 + n_3}{N} = \frac{1 + 1}{2000} = \frac{2}{2000} = 1{,}0 \times 10^{-3}$$

#### 4. Probabilidade de Falha do Sistema ($Q_{\text{sys}}$)

Para o sistema paralelo 1oo3, o sistema **falha somente quando todos os 3 componentes falham**.

As contribuições para $Q_{\text{sys}}$ são:

| Mecanismo | Expressão | Valor |
|-----------|-----------|-------|
| CCF de 3 componentes | $n_3 / N$ | $5{,}0 \times 10^{-4}$ |
| CCF de 2 + 1 independente | $3 \cdot \frac{n_2}{C(3,2) \cdot N} \cdot Q_i$ | $\approx 3{,}75 \times 10^{-6}$ |
| 3 falhas independentes | $Q_i^3$ | $\approx 4{,}2 \times 10^{-7}$ |

A contribuição dominante é a falha de causa comum tripla:

$$\boxed{Q_{\text{sys}} \approx \frac{n_3}{N} = \frac{1}{2000} = 5{,}0 \times 10^{-4}}$$

```python
from math import comb

# --- Dados do problema ---
m = 3        # número de componentes
N = 2000     # demandas totais (horas de missão)
n1 = 45      # eventos com 1 componente em falha
n2 = 1       # eventos com 2 componentes em falha simultânea
n3 = 1       # eventos com 3 componentes em falha simultânea

# --- Cálculos ---
total_falhas_componente = 1*n1 + 2*n2 + 3*n3

# 1. Probabilidade total de falha de um componente
Qt = total_falhas_componente / (m * N)

# 2. Probabilidade de falha independente de um componente
Qi = n1 / (m * N)

# 3. Probabilidade de falha simultânea de mais de um componente (evento CCF)
Q_cce = (n2 + n3) / N

# 4. Probabilidade de falha do sistema (1oo3: falha quando todos os 3 falham)
Q_cce3      = n3 / N                                  # CCF tripla (dominante)
Q_cce2_plus = 3 * (n2 / (comb(3, 2) * N)) * Qi       # CCF dupla + 1 independente
Q_ind3      = Qi**3                                    # 3 falhas independentes
Q_sys       = Q_cce3 + Q_cce2_plus + Q_ind3

# --- Resultados ---
print("=" * 55)
print("  Questão 8 — Sistema Paralelo 1oo3 com CCF")
print("=" * 55)
print(f"\n  Dados: m={m}, N={N}, n1={n1}, n2={n2}, n3={n3}")
print(f"  Total de falhas de componentes: {total_falhas_componente}\n")
print(f"  1. Qt  (prob. total de falha do componente)  = {Qt:.4e}")
print(f"  2. Qi  (prob. de falha independente)         = {Qi:.4e}")
print(f"  3. Qcce (prob. de falha de >1 simultaneamente)= {Q_cce:.4e}")
print(f"\n  4. Qsys (prob. de falha do sistema):")
print(f"       CCF tripla              = {Q_cce3:.4e}  (dominante)")
print(f"       CCF dupla + independente= {Q_cce2_plus:.4e}")
print(f"       3 independentes         = {Q_ind3:.4e}")
print(f"       ----------------------------------------")
print(f"       Q_sys TOTAL             = {Q_sys:.4e}")

# Verificação: Qt = Qi + contribuição CCF por componente
Q_ccf_comp = (2*n2/3 + n3) / N
print(f"\n  Verificação: Qt = Qi + Q_ccf/comp")
print(f"    {Qi:.4e} + {Q_ccf_comp:.4e} = {Qi + Q_ccf_comp:.4e}  (Qt = {Qt:.4e}) ✓")
```
