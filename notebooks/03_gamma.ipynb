## 1. Purpose

This notebook analyzes the Gamma distribution in the context of reliability engineering.

The objective is to:
- Compute the probability density function (PDF) and cumulative distribution function (CDF)
- Evaluate probabilities such as $P(X < x)$
- Understand its role in modeling waiting times and aggregated failure processes

The Gamma distribution generalizes the exponential distribution and is useful when modeling systems with multiple failure stages.

## 2. Mathematical Model

The Gamma distribution is defined by a shape parameter $\alpha$ and a scale parameter $\beta$.

The probability density function (PDF) is:

$$
f(x) = \frac{1}{\Gamma(\alpha)\,\beta^\alpha} x^{\alpha-1} e^{-x/\beta}, \quad x \geq 0
$$

The cumulative distribution function (CDF) is:

$$
F(x) = P(X < x) = \frac{\gamma(\alpha, x/\beta)}{\Gamma(\alpha)}
$$

where:
- $\Gamma(\alpha)$ is the Gamma function
- $\gamma(\alpha, x)$ is the lower incomplete Gamma function

The mean is:

$$
\mathbb{E}[X] = \alpha \beta
$$

```python
import numpy as np
from scipy.special import gamma, gammainc
import matplotlib.pyplot as plt

# Functions
def gamma_pdf(x, alpha, beta):
    return (x**(alpha-1) * np.exp(-x/beta)) / (gamma(alpha) * beta**alpha)

def gamma_cdf(x, alpha, beta):
    return gammainc(alpha, x/beta)
```

For a gamma distribution with a scale parameter of 400, and a shape parameter of 3.8, determine $Pr(x < 200)$.

```python
# Parameters
alpha = 3.8     # shape
beta = 400      # scale

# Calculation
x_value = 200
prob = gamma_cdf(x_value, alpha, beta)

print(f"P(X < {x_value}) = {prob:.5f}")
```
