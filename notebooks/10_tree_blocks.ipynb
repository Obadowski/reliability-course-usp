```python
# Question 4.3
import numpy as np

lam1 = 1e-5
lam2 = 1e-4
lam3 = 2e-5
lam4 = 5e-5

# Anonymous exponential function
exp_func = lambda t, lam: np.exp(-lam*t)

# Combined system
rsys = lambda t: exp_func(t, lam1+lam2) + exp_func(t, lam3+lam4) - exp_func(t, lam1+lam2+lam3+lam4)

# Reliability for 1000 hours, 5 decimal point precision
print(f"Reliability at 1000 hours: {rsys(1000):.5f}")

#! Attention, this topic uses numerical integration!
# MTTF calculation using numerical integration
from scipy.integrate import quad
mttf, _ = quad(rsys, 0, np.inf)
print(f"Mean Time To Failure (MTTF): {mttf:.2f} hours")

# Traditional calculation
MTTF = (1/(lam1+lam2)) + (1/(lam3+lam4)) - (1/(lam1+lam2+lam3+lam4))
print(f"Mean Time To Failure (MTTF) using traditional method: {MTTF:.2f} hours")
```

```python
# Question 4.3

from scipy.integrate import quad
import numpy as np

lv = 1e-3
lp = 2e-3
rsys_q403 = lambda t: np.exp(-lp*t) *(2*np.exp(-2*lv*t) - np.exp(-3*lv*t))

# Reliability at 10 and 20 hours
reliability_10_hours = rsys_q403(10)
print(f"Reliability at 10 hours: {reliability_10_hours:.5f}")
reliability_20_hours = rsys_q403(20)
print(f"Reliability at 20 hours: {reliability_20_hours:.5f}")

# Pr (T > 20 | T > 10) = Pr (T > 20) / Pr (T > 10)
conditional_reliability = reliability_20_hours / reliability_10_hours
print(f"Conditional reliability (T > 20 | T > 10): {conditional_reliability:.5f}")
```
