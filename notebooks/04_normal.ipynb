## Confidence interval

The confidence interval for the mean of a normal distribution are given by:

$$
Pr \left [ \hat{\mu}-t_{1-\alpha}(n-1)\times\frac{S}{\sqrt{n}} \leq \mu \leq  \hat{\mu}+t_{1-\alpha}(n-1)\times\frac{S}{\sqrt{n}} \right ] = 1 - \alpha
$$

Where,
1. $\hat{\mu}$ is the **estimated** mean
2. $t_{1-\alpha}(r-1)$ is the t-student for a confidence of $(1-\alpha)$ with $(r-1)$ degrees of freedom
3. $\alpha$ is the significance level
4. $r$ is the number of samples
5. $S$ is the **estimated** standard deviation

While, the confidence interval for the variance is:
$$
Pr \left [ \frac{\hat{\sigma}^2(n-1)}{\chi_{1-\alpha/2}^2(n-1)} \leq \sigma^2 \leq \frac{\hat{\sigma}^2(n-1)}{\chi_{\alpha/2}^2(n-1)}\right ] = 1 - \alpha
$$

Where,
1. $\hat{\sigma}^2$ is the **estimated** variance
2. $\chi_{\alpha/2}^2(n-1)$ is the chi-square distribution with confidence interval of $\alpha/2$ with $(n-1)$ degrees of freedom

## Exercise 3.15

The shaft diameters in a sample of 25 shafts are measured. The sample mean of diameter 
is 0.102 cm, with a standard deviation of 0.005 cm. What is the upper 95% confidence limit 
on the mean diameter of all shafts produced by this process, assuming that the distribution 
of shaft diameters is normal?

### Notes about the question

The question ask only for the upper limit, but it does not state whether it seeks out for a one-sided or two-sided analysis. Academically, both solution are provided, however I do use only the two-sided as it makes more sense in general.

```python
# One-sided solution
import math
from scipy.stats import t

# Data
xbar = 0.102
s = 0.005
n = 25
alpha = 0.05
df = n - 1

# Standard error
se = s / math.sqrt(n)

# One-sided upper t critical value
t_crit = t.ppf(1 - alpha, df)

# Upper confidence limit
upper_limit = xbar + t_crit * se

print(f"t critical value = {t_crit:.4f}")
print(f"Upper 95% confidence limit = {upper_limit:.4f} cm")
```

```python
# Two-sided solution

from scipy.stats import t
import math

alpha = 0.05
xbar = 0.102
s = 0.005
n = 25
df = n - 1
se = s / math.sqrt(n)

lower, upper = t.interval(1-alpha, df, loc=xbar, scale=se)
print(f"Lower 95% confidence limit = {lower:.4f} cm")
print(f"Upper 95% confidence limit = {upper:.4f} cm")
```

## Exercise 3.16

The sample mean life of 10 car batteries is 102.5 months, with a standard deviation of 9.45 
months. What are the 80% confidence limits for the mean and standard deviation of a pdf 
that represents these batteries?

```python
# two-sided solution
from scipy.stats import t, chi2
import math

alpha = 0.20
xbar = 102.5
s = 9.45
n = 10
se = s / math.sqrt(n)
df = n - 1

# Confidence interval for the mean
lower, upper = t.interval(1-alpha, df, loc=xbar, scale=se)
print(f"Lower 80% confidence limit for mean = {lower:.2f} months")
print(f"Upper 80% confidence limit for mean = {upper:.2f} months")

# Simply separation for ease of reading
print(f"----------------------------------")

# Confidence interval for the variance
chi2_lower = chi2.ppf(alpha / 2, df)
chi2_upper = chi2.ppf(1 - alpha / 2, df)
var_lower = (df * s**2) / chi2_upper
var_upper = (df * s**2) / chi2_lower

# Recalling that the limits are calculated for the variance,
# It is necessary to calculate the square root to obtain the confidence limits for the standard deviation.
std_lower = math.sqrt(var_lower)
std_upper = math.sqrt(var_upper)

print(f"Lower 80% confidence limit for std dev = {std_lower:.2f} months")
print(f"Upper 80% confidence limit for std dev = {std_upper:.2f} months")
```

## Exercise 3.17

The breaking strength X of five specimens of a rope of 1/4-inch diameter are 660, 460, 540, 
580, and 550 lb. Estimate the following:

a. The mean breaking strength by a 95% confidence level assuming normally distributed 
strength.

b. The point estimate of strength value at which only 5% of such specimens would be 
expected to break if μ is assumed to be an unbiased estimate of the true mean and σ is 
assumed to be the true standard deviation. (Assume x is normally distributed.)

c. The 90% confidence interval of the estimate of the standard deviation.

```python
# Solution

# Item a

import numpy as np
import pandas as pd
from scipy.stats import t

data = np.array([660, 460, 540, 580, 550])
xbar = np.mean(data)
s = np.std(data, ddof=1) # n-1 for sample standard deviation, as mean is estimated as well
n = len(data)
alpha = 0.05
df = n - 1
se = s / np.sqrt(n)

# Print additional info calculated for solution
print(f"Sample mean (xbar) = {xbar:.2f} lbs")
print(f"Sample standard deviation (s) = {s:.2f} lbs")
print(f"Sample size (n) = {n}")
print(f"Degrees of freedom (df) = {df}")
print(f"Standard error (se) = {se:.2f} lbs")

lower, upper = t.interval(1-alpha, df, loc=xbar, scale=se)
print(f"Lower 95% confidence limit = {lower:.2f} lbs")
print(f"Upper 95% confidence limit = {upper:.2f} lbs")
```

```python
# Item b

# If the values of xbar and std dev are the true values, then \phi(0.05) would be the probability where only 5% of the specimens, thus:

from scipy.stats import norm

critical_alpha = 0.05
z_crit = norm.ppf(critical_alpha)
print(f"Critical z value for alpha = {critical_alpha} is {z_crit:.4f}")

# Thus, given the xbar as st dev obtained as true values, one can expected that X_0.05 is:

x_critical = xbar + z_crit * s
print(f"Critical value X_0.05 = {x_critical:.2f} lbs")
```

```python
# Item c
# Standard deviation interval calculation for 90 % confidence level
from scipy.stats import chi2
import math
alpha = 0.10
df = n - 1
var_lower = (df * s**2) / chi2.ppf(1 - alpha / 2, df)
var_upper = (df * s**2) / chi2.ppf(alpha / 2, df)
std_lower = math.sqrt(var_lower)
std_upper = math.sqrt(var_upper)
print(f"Lower 90% confidence limit for std dev = {std_lower:.2f} lbs")
print(f"Upper 90% confidence limit for std dev = {std_upper:.2f} lbs")
```

## Exercise 3.20

The average life of a certain type of small motor is 10 years, with a standard deviation of 
2 years. The manufacturer replaces free of charge all motors that fail while under warranty. 
If the manufacturer is willing to replace only 3% of the motors that fail, what warranty 
period should be offered? Assume that the time to failure of the motors follows a normal 
distribution.

```python
# Exercise 20
# mean is 10 years, st dev is 2 years, warranty time assuming the manufacturer is willing to replace 3% of the products, thus:
from scipy.stats import norm
mean = 10
std_dev = 2
alpha = 0.03
z_crit = norm.ppf(alpha)
warranty_time = mean + z_crit * std_dev
print(f"Critical z value for alpha = {alpha} is {z_crit:.4f}")
print(f"Warranty time for 3% replacement = {warranty_time:.2f} years")
```

## Exercise 3.21

A manufacturer claims that certain machine parts will have a mean diameter of 4 cm, with 
a standard deviation of 0.01 mm. The diameters of five parts are measured and found to 
be (in mm) 39.98, 40.01, 39.96, 40.03, and 40.02. Would you accept this claim with a 90% 
confidence level?

```python
# Exercise 3.21

# mean diameter 4cm
# st dev 0.01 mm 
# Five parts produced with 39.98, 40.01, 39.96, 40.03 and 40.02
# Would you accept at 90% confidence level?

from scipy.stats import norm, t, chi2
import numpy as np

# data from the exercise
data = np.array([39.98, 40.01, 39.96, 40.03, 40.02])
mean = 40.00
std_dev = 0.01
alpha = 0.10

df = len(data) - 1          # degrees of freedom for t-distribution
xbar = np.mean(data)        # sample mean
s = np.std(data, ddof=1)    # sample standard deviation (ddof=1 for sample standard deviation)
se = s / np.sqrt(len(data)) # standard error of the mean

# Sample mean confidence interval
t_crit = t.ppf(1 - alpha / 2, df)   # two-tailed t critical value for 90% confidence
lower_limit = xbar - t_crit * se    # lower confidence limit
upper_limit = xbar + t_crit * se    # upper confidence limit

# Standard error confidence interval
chi2_lower = chi2.ppf(alpha / 2, df)   # lower chi
chi2_upper = chi2.ppf(1 - alpha / 2, df) # upper chi
var_lower = (df * s**2) / chi2_upper   # lower confidence limit for variance
var_upper = (df * s**2) / chi2_lower   # upper confidence limit for variance
std_lower = np.sqrt(var_lower)         # lower confidence limit for standard deviation
std_upper = np.sqrt(var_upper)         # upper confidence limit for standard deviation

print(f"Sample mean (xbar) = {xbar:.4f} cm")
print(f"Sample standard deviation (s) = {s:.4f} cm")
print(f"Standard error (se) = {se:.4f} cm")
print(f"t critical value for 90% confidence = {t_crit:.4f}")

# Evaluation of mean proposed by the manufacturer
print(f"--"*30)
print(f"Proposed mean = {mean:.4f} cm")
print(f"90% confidence interval for the mean = ({lower_limit:.4f}, {upper_limit:.4f}) cm")
if lower_limit <= mean <= upper_limit:
    print("Mean is within the interval")
else:    
    print("(!!)Mean is not within the interval")

# Evaluation of standard deviation proposed by the manufacturer
print(f"--"*30)
print(f"Proposed standard deviation = {std_dev:.4f} cm")
print(f"90% confidence interval for the standard deviation = ({std_lower:.4f}, {std_upper:.4f}) cm")
if std_lower <= std_dev <= std_upper:
    print("Standard deviation is within the interval")
else:
    print("(!!)Standard deviation is not within the interval")
```

## Exercise 3.25

A sample of 10 measurements of a sphere diameter gives a mean of 4.38 inches, with a 
standard deviation of 0.06 inch. Find the 99% confidence limits of the actual mean and 
standard deviation.

```python
# Exercise 3.25
# mean = 4.38 inches for 10 measurements
# st dev = 0.06 inch
# 99% confidence interval for the mean and standard deviation

from scipy.stats import t, chi2
import math

x_hat = 4.38  # mean
s = 0.06  # standard deviation
n = 10  # number of measurements
alpha = 0.01  # significance level for 99% confidence
df = n - 1  # degrees of freedom
se = s / math.sqrt(n)  # standard error of the mean
# Confidence interval for the mean
lower_mean, upper_mean = t.interval(1 - alpha, df, loc=x_hat, scale=se)
print(f"99% confidence interval for the mean = ({lower_mean:.4f}, {upper_mean:.4f}) inches")
# Confidence interval for the standard deviation
chi2_lower = chi2.ppf(alpha / 2, df)
chi2_upper = chi2.ppf(1 - alpha / 2, df)
var_lower = (df * s**2) / chi2_upper
var_upper = (df * s**2) / chi2_lower
std_lower = math.sqrt(var_lower)
std_upper = math.sqrt(var_upper)
print(f"99% confidence interval for the standard deviation = ({std_lower:.4f}, {std_upper:.4f}) inches")
```
