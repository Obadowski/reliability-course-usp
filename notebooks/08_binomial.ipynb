## Maximum Likelihood Estimator

$$
\hat{p} = \frac{x}{n}
$$

Where:
1. $\hat{p}$ is the point estimation for the failure rate of binominal distribution (or occurence.)
2. $x$ is the number of events (failure, for instance)
3. $n$ is the number of tentatives

The lower and upper confidence intervals for a confidence level of $\alpha$ is:

$$
p_{lower} = \{ 1 + (n - x + 1) x^{-1} F_{1-\alpha/2}[2n-2x+2;2x] \}^{-1}
$$


$$
p_{upper} = \{ 1 + (n - x)\{ (x + 1) F_{1-\alpha/2}[2x+2;2n-2x]\}^{-1}\}^{-1}
$$

Where $F_{1-\alpha/2}[\nu_1;\nu_2]$ where $\nu_1$ is the degrees of freedom of numerator and $\nu_2$ degrees of freedom of denominator.

## Exercise 3.24

A locomotive control system fails 15 times out of the 96 times it is activated to function.  Determine 
the following: 

a. A point estimate for failure probability of the system 

b. 95% two-sided confidence intervals for the probability of failure. (Assume that after each 
failure, the system is repaired and put back in an as-good-as-new state.) 

```python
# Binominal distribution calculator
# 15 failures out of 96
# point estimation of the probability of failure
# confidence interval for the probability of failure using Clopper-Pearson

from scipy.stats import f

x = 15   # failures
n = 96   # tentatives

p_hat = x / n
alpha = 0.05
print(f"Point estimation of the probability of failure: {p_hat:.4f}")

f_value = f.ppf(1-alpha/2, 2*(n-x+1), 2*(x))
print(f"alpha, nu1, nu2: {alpha}, {2*x}, {2*(n-x+1)}")
print(f"F value for upper bound: {f_value:.4f}")

p_lower = (1 + (n - x + 1) * (x**(-1)) * f.ppf(1-alpha/2, 2*(n-x+1), 2*x)) ** (-1)
print(f"Lower bound of the confidence interval: {p_lower:.4f}")

internal = ((x + 1) * f.ppf(1-alpha/2, 2*(x+1), 2*(n-x)))**(-1)
p_upper = (1 + (n - x) * internal) ** (-1)

print(f"Upper bound of the confidence interval: {p_upper:.4f}")
```

```python
# Binomial distribution calculator

# Exercise 4.14 from Modarres, 30 filters, each one exponential with lambda = 2e-4, fails if 3 or more filters fail
from scipy.stats import binom
import numpy as np

l = 2e-4
Fail_single_filter = 1 - np.exp(-l*1000)  # failure probability of a single filter in 1000 hours
print(f"Failure probability of a single filter in 1000 hours: {Fail_single_filter:.6f}")

n_filters = 30
k_failures = 3
prob_failure = 1 - binom.cdf(k_failures - 1, n_filters, Fail_single_filter)
print(f"Probability of failure (3 or more filters fail): {prob_failure:.6f}")

# Reliability
print(f"Reliability (less than 3 filters fail): {1 - prob_failure:.6f}")
```
